```python
from tkinter import*

def process():                             #함수 process() 정의
    temperature = float(e1.get())         #변수 temperature 선언, 실수형으로 받기
    mytemp = (temperature-32)*5/9         #화씨에서 섭씨로 변경하는 식
    e2.insert(0, str(mytemp))

def process2():                                #함수 process2() 정의
    temperature = float(e2.get())             #변수 temperature 선언, 실수형으로 받기
    mytemp = (temperature*(9/5))+32          #섭씨에서 화씨로 변경하는 식
    e1.insert(0, str(mytemp))

window = Tk()

l1 = Label(window, text = "화씨")
l2 = Label(window, text = "섭씨")
l1.grid(row = 0, column = 0)
l2.grid(row = 1, column = 0)

e1 = Entry(window)
e2 = Entry(window)
e1.grid(row = 0, column = 1)
e2.grid(row = 1, column = 1)

b1 = Button(window, text = "화씨 -> 섭씨", command=process)         #화씨에서 섭씨로 바꾸도록 하는 버튼 위젯 생성
b2 = Button(window, text = "섭씨 -> 화씨", command=process2)         #섭씨에서 화씨로 바꾸도록 하는 버튼 위젯 생성
b1.grid(row = 2, column = 0)
b2.grid(row = 2, column = 1)

window.mainloop()
```

```python
from tkinter import*
def paint(event):                                          #함수 paint() 정의
    global lastx, lasty                                    #global;전역변수
    x, y = (event.x),(event.y)
    canvas.create_line(lastx,lasty,x,y,fill = "black")
    lastx, lasty = x, y 

def activate_paint(event):                      #함수 activate_paint() 정의
    global lastx, lasty
    lastx, lasty = (event.x),(event.y)

def release(event):                               #함수 release() 정의
    global lastx, lasty 

    if (lastx, lasty) == (event.x, event.y):
        canvas.create_line(lastx,lasty,lastx+1,lasty+1)

lastx, lasty = None, None

window = Tk()
canvas = Canvas(window)
canvas.pack()
canvas.bind('<B1-Motion>', paint)                 #위젯.bind(): 이벤트 처리 함수      #드래그 할 때 이벤트 발생
canvas.bind('<ButtonPress-1>', activate_paint)    #클릭할 때 이벤트 발생
canvas.bind('<ButtonRelease-1>', release)         #떼었을 때 이벤트 발생
window.mainloop()                                 #mainioop(): 이벤트를 기다리는 함수
```

```python
from tkinter import*
mycolor = "blue"
def paint(event):                                          #함수 paint() 정의
    x1, y1 = (event.x-1), (event.y+1)
    x2, y2 = (event.x-1), (event.y+1)
    canvas.create_oval(x1, y1, x2, y2, fill = mycolor)

def change_color_red():                        #빨간색으로 바꾸는 함수 change_color_red() 정의
    global mycolor
    mycolor = "red"

def change_color_green():                      #녹색으로 바꾸는 함수 change_color_green() 정의
    global mycolor
    mycolor = "green"

def change_color_yellow():                     #노란색으로 바꾸는 함수 change_color_yellow() 정의
    global mycolor
    mycolor = "yellow"

window = Tk()
canvas = Canvas(window)
canvas.pack()
canvas.bind("<B1-Motion>", paint)
button = Button(window, text="빨간색", command=change_color_red)          #빨간색으로 바꾸는 버튼 위젯 만들기
button.pack()
button = Button(window, text="녹색", command=change_color_green)           #녹색으로 바꾸는 버튼 위젯 만들기
button.pack()
button = Button(window, text="노란색", command=change_color_yellow)    #노란색으로 바꾸는 버튼 위젯 만들기
button.pack()
window.mainloop()
```

```python
from tkinter import Tk, Label, Button, Entry, IntVar, END, W, E 

def update_add():                      #함수 update_add() 정의
    update("add")

def update_subtract():                 #함수 update_subtract() 정의
    update("subtract")

def update_reset():                    #함수 update_reset() 정의
    update("reset")

window = Tk()
total = 0                              #0으로 초기화
sum = Label(window)
sum.grid(row=0, column=1, columnspan=2)

label = Label(window, text="현재 합계: ")
label.grid(row=0, column=0)

entry = Entry(window)
entry.grid(row=1, column=0, columnspan=3)

add_button = Button(window, text="더하기(+)", command=update_add)             #더하는 버튼 위젯 생성
subtract_button = Button(window, text="빼기(-)", command=update_subtract)     #빼는 버튼 위젯 생성
reset_button = Button(window, text="초기화", command=update_reset)            #리셋, 초기화하는 버튼 위젯 생성

add_button.grid(row=2, column=0)
subtract_button.grid(row=2, column=1)
reset_button.grid(row=2, column=1)

def update(method):                                 #함수 update() 정의
    global total
    if method == "add":                             #더하는 경우
        total += int(entry.get())
    elif method == "subtract":                      #빼는 경우
        total -= int(entry.get())
    else:                                           #그 외의 경우(리셋)
        total = 0                                   #0으로 초기화
    sum['text'] = str(total)
    entry.delete(0, END)
window.mainloop()
```

```python
import random                             #랜덤함수, 난수모듈 불러오기
from tkinter import*

window = Tk()
secret_number = random.randint(1, 100)    #주어진 범위 중에서 랜덤 선택
guess = None
num_guesses = 0                           #0으로 초기화

def guess_number():                       #함수 guess_number() 정의
    global num_guesses
    guess = int(entry.get())              #변수 guess 선언, 정수형으로 받기
    num_guesses += 1
    if guess == secret_number:            #추측이 랜덤 선택된 수와 일치한다면
        message = "축하합니다!"           #축하합니다! 출력
    elif guess < secret_number:           #추측이 랜덤 선택된 수보다 작다면
        message = "너무 낮아요!"          #너무 낮아요! 출력
    else:                                 #그 외의 경우(추측이 랜덤 선택된 수보다 크다면)
        message = "너무 높아요!"          #너무 높아요! 출력
    label['text'] = message

def reset():                              #함수 reset() 정의
    global num_guesses
    entry.delete(0, END)
    secret_number = random.randint(0, 100)               #주어진 범위 중에서 랜덤 선택
    guess = 0                                            #0으로 초기화
    message = "1부터 100까지의 숫자를 추측하시오."       
    label['text'] = message

message = "1부터 100 사이의 숫자를 추측하시오."
label = Label(window, text=message)
entry = Entry(window)

guess_button = Button(window, text="숫자를 입력", command=guess_number)       #추측한 숫자를 기입하고 누르는 버튼 위젯 생성
reset_button = Button(window, text="게임을 다시 실행", command=reset)         #리셋하는 버튼 위젯 생성

label.grid(row=0, column=0, columnspan=2, sticky=W+E)
entry.grid(row=1, column=0, columnspan=2, stickey=W+E)
guess_button.grid(row=2, column=0)
reset_button.grid(row=2, column=1)

window.mainloop()
```

```python
from tkinter import*

def do_convert():                                            #함수 do_convert() 정의
    inch_val = float(cvt_from.get())
    centimeters_val = inch_val * 2.54
    cvt_to.set('%s 센티미터' % centimeters_val)

root = Tk()
cvt_from = StringVar()
cvt_to = StringVar()

lbl = Label(root, text="인치를 센티미터로 변환하는 프로그램")
lbl.grid(row=0, column=0, columnspan=2)
from_lbl = Label(root, text='인치를 입력하시오:')
from_lbl.grid(row=1, column=0)
from_entry = Entry(root, textvariable=cvt_from)
from_entry.grid(row=1, column=1)

to_lbl = Label(root, text='변환결과: ')
to_lbl.grid(row=2, column=0)
result_lbl = Label(root, textvariable=cvt_to)
result_lbl.grid(row=2, column=1)

convert_btn = Button(root, text='변환!', command=do_convert)          #변환하는 버튼 위젯 생성
convert_btn.grid(row=3, column=1)

root.mainloop()
```

```python
from tkinter import*
fields = '이름', '직업', '국적'

def fetch(entries):                                   #함수 fetch() 정의
    for entry in entries:                             #반복 함수
        field = entry[0]
        text = entry[1].get()
        print('%s: "%s"' % (field, text))

def makeform(root, fields):                                   #함수 makeform() 정의
    entries = []
    for field in fields:                                  #반복 함수
        row = Frame(root)
        lab = Label(row, width=15, text=field)
        ent = Entry(row)
        row.pack(side=TOP, fill=X)
        lab.pack(side=LEFT)
        ent.pack(side=RIGHT, expand=YES, fill=X)
        entries.append((field, ent))
    return entries

root = Tk()
ents = makeform(root, fields)
root.bind('<Return>', (lambda event, e=ents: fetch(e)))
b1 = Button(root, text='보여주기', command=(lambda e=ents: fetch(e)))          #보여주기 버튼 위젯 생성
b1.pack(side=LEFT, padx=5, pady=5)
b2 = Button(root, text='종료하기', command=root.quit)                          #종료하기 버튼 위젯 생성
b2.pack(side=LEFT, padx=5, pady=5)
root.mainloop()
```

```python
from random import*
from tkinter import*

def user_choice_rock():                            #함수 user_choice_rock() 정의
    user_choice = "rock"                           #사용자가 주먹 냄
    turn(user_choice)
    user_image.configure(image=rock_image)         

def user_choice_paper():                            #함수 user_choice_paper() 정의
    user_choice = "paper"                           #사용자가 보 냄
    turn(user_choice)
    user_image.configure(image=paper_image)

def user_choice_scissors():                          #함수 user_choice_scissors() 정의
    user_choice = "scissors"                         #사용자가 가위 냄
    turn(user_choice)
    user_image.configure(image=scissors_image)

#게임
def turn(user_choice):                               #함수 turn() 정의
    oppo = ['rock', 'paper', 'scissors']
    oppo_choice=oppo[randint(0, 2)]                  #랜덤 선택하기
    if(oppo_choice=='rock'):                          #랜덤 선택한 것이 주먹이라면
        oppo_image.configure(image=rock_image)         #주먹 이미지
        if(user_choice=='paper'):                                           #사용자가 보를 냈다면 
            turn_result.configure(text='사용자 승', fg="green")             #사용자가 승
            compare.configure(text=">>>>>")
        elif(user_choice=='scissors'):                                        #사용자가 가위를 냈다면
            turn_result.configure(text="컴퓨터 승!", fg="red")                #컴퓨터가 승
            compare.configure(text="<<<<<")
        else:                                                                  #그 밖의 경우(둘 다 주먹)
            turn_result.configure(text="무승부", fg="gray")                    #무승부
            compare.configure(text="=====")
    elif(oppo_choice=='paper'):                                                #랜덤 선택한 것이 보라면
        oppo_image.configure(image=paper_image)                                #보 이미지
        if(user_choice=='scissors'):                                           #사용자가 가위를 냈다면
            turn_result.configure(text="사용자 승!", fg="green")                #사용자 승
            compare.configure(text="사용자가 승리하였습니다.")               
        elif(user_choice=='rock'):                                                  #사용자가 주먹을 냈다면
            turn_result.configure(text="이런! 컴퓨터가 이겼습니다!", fg="red")      #컴퓨터 승
            compare.configure(text="<<<<<")
        else:                                                                      #그 밖의 경우(둘 다 보)
            turn_result.configure(text="무승부", fg="gray")                        #무승부
            compare.configure(text="=====")
    
    elif(oppo_choice=='scissors'):                                                  #랜덤 선택한 것이 가위라면
        oppo_image.configure(image=scissors_image)                                  #가위 이미지
        if(user_choice=='rock'):                                                     #사용자가 주먹을 냈다면
            turn_result.configure(text="사용자가 이겼습니다!", fg="green")           #사용자 승
            compare.configure(text=">>>>>")                  
        elif(user_choice=='paper'):                                                 #사용자가 보를 냈다면
            turn_result.configure(text="에고 컴퓨터가 이겼군요!", fg="gray")        #컴퓨터 승
            compare.configure(text="<<<<<")
        else:                                                                       #그 밖의 경우(둘 다 가위)
            turn_result.configure(text="무승부", fg="gray")                         #무승부
            compare.configure(text="=====")

#메인 프로그램 부분
main_window = Tk()
rock_button = Button(main_window, width=20, text="바위", justify=CENTER, command=user_choice_rock, activebackground='black', activeforeground='white')         #바위 내는 버튼 위젯 생성

paper_button = Button(main_window, width=20, text="보", justify=CENTER, command=user_choice_paper, activebackground='black', activeforeground='white')        #보 내는 버튼 위젯 생성

scissors_button = Button(main_window, width=20, text="가위", justify=CENTER, command=user_choice_scissors, activebackground='black', activeforeground='white')     #가위 내는 버튼 위젯 생성

rock_image = PhotoImage(file="rock.gif")            #바위 이미지 파일에서 불러오기
paper_image = PhotoImage(file="paper.gif")            #보 이미지 파일에서 불러오기
scissors_image = PhotoImage(file="scissors.gif")     #가위 이미지 파일에서 불러오기

user_image = Label(text="사용자", image=rock_image)
user_image.image = rock_image

compare = Label(main_window, justify=CENTER, font=("Helvetica", 20))
oppo_image = Label(text="컴퓨터", image=paper_image)
oppo_image.image = paper_image
turn_result = Label(main_window, width=20, justify=CENTER, font=("Helvetica", 20))

#그리드 생성
rock_button.grid(row=5, column=1)
paper_button.grid(row=5, column=2)
scissors_button.grid(row=5, column=3)
user_image.grid(row=3, column=1)
compare.grid(row=3, column=2)
oppo_image.grid(row=3, column=3)
turn_result.grid(row=4, column=2)

#메인 루프 처리(GUI 화면을 띄운 상태로)
main_window.mainloop()
```
